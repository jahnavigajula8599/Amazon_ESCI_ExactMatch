```python
df_joined = df_examples.merge(
    df_products,
    how="left",
    on=["product_id","product_locale"]
)

# Filter where esci_label == "E" (Exact matches)
df_E = df_joined[df_joined["esci_label"] == "E"]
```

```python
# Select only the requested columns
cols_to_show = [
    "query", "product_id", "product_locale", 
    "product_title", "product_description", 
    "product_bullet_point", "product_brand", "product_color"
]
```

```python
import csv

df_E[cols_to_show].head(100).to_csv(
    "esci_E_exact_matches_sample100.csv",
    index=False,
    quoting=csv.QUOTE_ALL   # ensure every field is quoted
)
```
